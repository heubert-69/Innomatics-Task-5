```python
#Problem #1
def analyze_post_engagement(likes):
    total_likes = sum(likes)
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    return f"Total Likes: {total_likes}\nPost Status: {status}"
```

```python
#Problem #2
def check_medicine_stock(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"
    return f"Medicine Stock: {stock}\nStatus: {status}"
```

```python
#Problem #3
def check_rainfall_adequacy(rainfall_data, required_level):
    avg_rainfall = sum(rainfall_data) / len(rainfall_data)
    status = "Adequate Rainfall" if avg_rainfall >= required_level else "Inadequate Rainfall"
    return f"Average Rainfall: {avg_rainfall}\nRainfall Status: {status}"
```

```python
#Problem #4
def detect_duplicate_usernames(usernames):
    has_duplicates = len(usernames) != len(set(usernames))
    result = "Yes" if has_duplicates else "No"
    return f"Duplicate Accounts Found: {result}"
```

```python
#Problem #5
def check_appointment_eligibility(age):
    status = "Eligible" if age >= 18 else "Not Eligible"
    return f"Patient Age: {age}\nEligibility Status: {status}"
```

```python
#Problem #6
def filter_premium_crops(prices, threshold=2000):
    premium_crops = [price for price in prices if price > threshold]
    return f"Premium Crops: {premium_crops}"
```

```python
#Problem #7
def check_system_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    return f"Error Count: {error_count}\nSystem Status: {status}"
```

```python
#Problem #8
def validate_transaction(amount, daily_limit=50000):
    status = "Approved" if amount <= daily_limit else "Rejected"
    return f"Transaction Amount: {amount}\nTransaction Status: {status}"
```

```python
#Problem #9
def check_exam_eligibility(attendance_list, total_classes):
    attended = sum(attendance_list)
    percentage = (attended / total_classes) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    return f"Attendance Percentage: {percentage}\nExam Eligibility: {status}"
```

```python
#Problem #10
def calculate_electricity_bill(units):
    # Calculate bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Classify usage
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return f"Bill Amount: ₹{bill}\nUsage Status: {usage}"
```
